```python
data['Route']=data['Route'].fillna(method='bfill')
data['Total_Stops']=data['Total_Stops'].fillna(method='bfill')

```

```python
data.head()
```

```python
data['Dep_day']=pd.to_datetime(data['Date_of_Journey'],format='%d/%m/%Y').dt.day
data['Dep_month']=pd.to_datetime(data['Date_of_Journey'],format='%d/%m/%Y').dt.month
data['Dep_year']=pd.to_datetime(data['Date_of_Journey'],format='%d/%m/%Y').dt.year
data
```

```python
data['Dep_hour']=pd.to_datetime(data['Dep_Time']).dt.hour
data['Dep_minute']=pd.to_datetime(data['Dep_Time']).dt.minute
data
```

```python
data.head(5)
```

```python
data.drop(['Date_of_Journey'],axis=1,inplace=True)
data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Arrival_Time'].value_counts()
```

```python
arr_h=[]
arr_m=[]
for a in data['Arrival_Time']:
    # print(a)
    a=a[:5]
    d=a.split(':')
    
    arr_h.append(d[0])
    arr_m.append(d[1])
```

```python
data['Arrival_hour']=arr_h
data['Arrival_minute']=arr_m
```

```python
minute=[]
hour=[]
for t in data['Duration']:
    time=t.split()
    print(time)
    try:
        h=time[0].replace('h','')
        m=time[1].replace('m','')
    except:
        h=time[0].replace('h','')
        m=0


    hour.append(h)
    minute.append(m)
```

```python
data['dur_h']=hour
data['dur_m']=minute
```
